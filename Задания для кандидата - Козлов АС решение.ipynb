## Задание 1

#### Простая задача, можно решить любым способом. Необходимо узнать является ли слово палиндромом без учета регистра (ШаЛАш - палиндром) и без учета пробелов (Ша Ла  ш - тоже палиндром). На вход подается строка, на выходе True или False.

```python
word = input()

def palindrom(word):
    w = word.replace(" ", "").upper()
    w_rev = w[::-1]
    print (w == w_rev)

palindrom(word)
```

```python
pr.loc[(pr.Класс == 'ТРЦ')|(pr.Класс == 'РТЦ')|(pr.Класс == 'Стрит-ритейл')] = pr.loc[(pr.Класс == 'ТРЦ')|(pr.Класс == 'РТЦ')|(pr.Класс == 'Стрит-ритейл')].fillna('Торговая недвижимость')
pr.loc[(pr.Класс == 'ОСН')] = pr.loc[(pr.Класс == 'ОСН')].fillna('Объекты свободного назначения')
pr.loc[(pr.Класс == 'Офисы A')|(pr.Класс == 'Офисы B')|(pr.Класс == 'Офисы C')] = pr.loc[(pr.Класс == 'Офисы A')|(pr.Класс == 'Офисы B')|(pr.Класс == 'Офисы C')].fillna('Офисная недвижимость')
pr.loc[(pr.Класс == 'ПС A')|(pr.Класс == 'ПС B')|(pr.Класс == 'ПС C')] = pr.loc[(pr.Класс == 'ПС A')|(pr.Класс == 'ПС B')|(pr.Класс == 'ПС C')].fillna('Производственно-складская нежвимость')
```
